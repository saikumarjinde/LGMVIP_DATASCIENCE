```python
import cv2
import matplotlib.pyplot as plt
```

# READ THE IMAGE

```python
image = cv2.imread("mountain.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,8))
plt.imshow(image)
plt.axis('off')
plt.title('THE ORIGINAL IMAGE')
plt.show()
```

# CONVERT IMAGE INTO GRAYSCALE

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,8))
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('THE GRAY IMAGE')
plt.show()
```

# GRAYSCALE INVERTED IMAGE

```python
inverted_image = 255 - gray_image
plt.figure(figsize=(10,8))
plt.imshow(inverted_image,cmap='gray')
plt.axis('off')
plt.title('THE INVERTED IMAGE')
plt.show()
```

# BLURRED IMAGE

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
plt.figure(figsize=(10,8))
plt.imshow(blurred_image,cmap='gray')
plt.axis('off')
plt.title('THE BLURRED INVERTED IMAGE')
plt.show()
```

# PENCIL SKETCH

```python
inverted_blurred = 255 - blurred_image
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=254.0)
plt.figure(figsize=(10,8))
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.title('THE PENCIL SKETCH')
plt.show()
```

# THE BLACK AND WHITE IMAGE

```python
from PIL import Image
import numpy as np
img = Image.open('mountain.jpg')
img_gray = img.convert('L')
img_array = np.array(img_gray)
plt.axis('off')
plt.title('THE BLACK AND WHITE SKETCH')
plt.imshow(img_array, cmap='gray')
```

# COMPARISON OF ORIGINAL,PENCIL AND BLACK AND WHITE SKETCH

```python
plt.figure(figsize=[10,5])
plt.subplot(1,3,1)
plt.imshow(image)
plt.axis('off')
plt.title('Original Sketch')
plt.subplot(1,3,3)
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.title('Pencile Sketch')
plt.subplot(1,3,2)
plt.imshow(img_array, cmap='gray')
plt.axis('off')
plt.title('Black and White Sketch')
```
